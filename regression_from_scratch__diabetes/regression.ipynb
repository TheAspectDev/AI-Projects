```python
import matplotlib.pyplot as plt 
import numpy as np    

def abline(slope, intercept):
    x = [1,2,3,4,5]
    y = [3, 5, 7, 9, 11]
    plt.plot(x,y, "yo")
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals 
    plt.plot(x_vals, y_vals, '--')
    
abline(2, 1)
```
