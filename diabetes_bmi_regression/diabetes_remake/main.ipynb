```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import sklearn.metrics as sm
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
plt.subplots(figsize=(9,9))
sns.heatmap(df.corr(), annot=True)
```

```python
x = df.drop("DiabetesPedigreeFunction", axis=1)
x = x.drop("Outcome", axis=1)
x = x.drop("BMI", axis=1)
x = x.drop("Insulin", axis=1)
# x = x.drop("BloodPressure", axis=1)
x = x.drop("Glucose", axis=1)
x = x.drop("Pregnancies", axis=1)

y = df.drop("DiabetesPedigreeFunction", axis=1)
y = y.drop("Age", axis=1)
y = y.drop("BMI", axis=1)
y = y.drop("Insulin", axis=1)
y = y.drop("SkinThickness", axis=1)
y = y.drop("BloodPressure", axis=1)
y = y.drop("Glucose", axis=1)
y = y.drop("Pregnancies", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)

X_test
```

```python
output = model.predict(X_test)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_test['Age'], X_test['SkinThickness'], y_test, color='blue', label='Data')
ax.scatter(X_test['Age'], X_test['SkinThickness'], output, color='red', label='Pred Data')

ax.set_xlabel('Age')
ax.set_ylabel('SkinThickness')
ax.set_zlabel('Diabetes Outcome')
ax.legend()

plt.show()
```

```python
age_skinthickness = np.array([[28, 23]])  # Age and SkinThickness
result = model.predict(age_skinthickness.reshape(1, -1))
print(result)
```
